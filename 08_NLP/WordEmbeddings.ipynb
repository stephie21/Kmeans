```python
port fasttext
import fasttext.util
fasttext.util.download_model('en', if_exists='ignore')  # English
```

2. How large is the vocabulary?

```python
model = fasttext.load_model('cc.en.300.bin')
```

```python
print("Vocab size:", len(model.words))

```

3. How many dimensions does the model have?

```python
print("Dimensions:", model.get_dimension())
```

4. Print the Word Vectors for the Munich!

```python
print("Vector for 'Munich':", model.get_word_vector("Munich")[:10])  # nur erste 10 Zahlen
```

5. What are the most similar terms to LRZ?

```python
print("Most similar to 'LRZ':", model.get_nearest_neighbors("LRZ")[:5])
```

6. Find the analogie to the relation of Berlin to Germany for Denmark!

```python
print("Berlin : Germany :: ? : Denmark →", model.get_analogies("Berlin", "Germany", "Denmark")[0][1])
```

Token Classification

```python
ner = pipeline("ner", grouped_entities=True)
ner("Albert Einstein was born in Ulm, Germany.")
```

Question Answering

```python
qa = pipeline("question-answering")
qa({
  "question": "Where was Einstein born?",
  "context": "Albert Einstein was born in Ulm, Germany."
})
```

Masked Language Modeling - filling missing word
